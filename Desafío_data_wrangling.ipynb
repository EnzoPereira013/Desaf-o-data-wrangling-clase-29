```python
df = df.drop(df.index[-1])
```

```python
df = df.rename(columns={"Product Category": "Category", "Sub Category": "Subcategory"})
```

```python
df = df.drop("index", axis=1)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
```

```python
df["Customer Gender"] = df["Customer Gender"].astype("category")
```

```python
column1_unique = df["Column1"].unique()
```

```python
print(column1_unique)
```

```python
media_column1 = df["Column1"].mean()
```

```python
df["Column1"].fillna(media_column1, inplace=True)
```

```python
duplicados = df.duplicated()
duplicados_df = df[duplicados]
print(duplicados_df)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year
df = df.drop("Date", axis=1)
```

```python
df
```

```python
categorical_columns = ["Customer Gender", "Country", "State", "Category", "Subcategory"]
df_encoded = pd.get_dummies(df, columns=categorical_columns)
```

```python
print(df_encoded.head())
```

```python
bike_columns = ["Subcategory_Mountain Bikes", "Subcategory_Road Bikes", "Subcategory_Touring Bikes"]
df_encoded["Subcategory_Bikes"] = df_encoded[bike_columns].sum(axis=1)
```

```python
df_encoded = df_encoded.drop(bike_columns, axis=1)
```

```python
print(df_encoded.head())
```

```python
numeric_columns = df_encoded.select_dtypes(include="number").columns
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
df_pca = pca.fit_transform(df_encoded)
df_pca = pd.DataFrame(df_pca, columns=["Componente 1", "Componente 2"])
print(df_pca.head())
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
print("Proporción de varianza explicada:", explained_variance_ratio)
```

```python
df_final = pd.concat([df_encoded.drop(numeric_columns, axis=1), pd.DataFrame(df_pca, columns=["Componente 1", "Componente 2"])], axis=1)
```
